## 3. Data Preparation

```python
df_modeling = df
mapping = {
    "Sehr Schlecht": 1, 
    "Sehr schlecht": 1, 
    "Schlecht": 2, 
    "Durchschnitt": 3, 
    "Gut": 4, 
    "Sehr gut": 5
}
df_modeling['Aussenqualitaet'] = df_modeling['Aussenqualitaet'].replace(mapping)
df_modeling['Aussenzustand'] = df_modeling['Aussenzustand'].replace(mapping)
df_modeling['Heizungszustand'] = df_modeling['Heizungszustand'].replace(mapping)
df_modeling['Kaminqualitaet'] = df_modeling['Kaminqualitaet'].replace(mapping)
df_modeling['Kuechenzustand'] = df_modeling['Kuechenzustand'].replace(mapping)
df_modeling['Kaminqualitaet'] = df_modeling['Kaminqualitaet'].astype('Int8')

mapping2 = {
    "Keine": -1,
    "Familienverkauf": 1, 
    "Pool": 2
}
df_modeling["Besonderheiten"] = df_modeling["Besonderheiten"].replace(mapping2)
df_modeling["Besonderheiten"] = df_modeling["Besonderheiten"].fillna(-1)
df_modeling['Besonderheiten'] = df_modeling['Besonderheiten'].astype('int8')

mapping3 = {
    "Keinen": -1,
    "Guter Wohnraum": 1, 
    "Rohbau": 2, 
    "Mittlerer Wohnraum": 3, 
    "Freizeitraum": 4, 
    "Kein Wohnraum": 5, 
    "Niedrige Qualität": 6
}
df_modeling["Keller"] = df_modeling["Keller"].replace(mapping3)
df_modeling['Keller'] = df_modeling['Keller'].astype('int8')

mapping4 = {
    "Upper East Side": 1, 
    "Lower East Side": 2, 
    "Queensland Doors": 3, 
    "Upper West Side": 4, 
    "Heroes Park": 5
}
df_modeling["Lage"] = df_modeling["Lage"].replace(mapping4)
df_modeling['Lage'] = df_modeling['Lage'].astype('Int8')

df_modeling["Kaminqualitaet"] = df_modeling["Kaminqualitaet"].fillna(-1)
df_modeling["Lage"] = df_modeling["Lage"].fillna(-1)

df_modeling.info()
```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict

def validation(model, x_train, y_train):   
    def neg_mean_absolute_percentage_error(model, X, y):
        y_pred = model.predict(X)
        return -np.mean(np.abs((y - y_pred) / y))
     
    return {"mean CV score": cross_val_score(model, x_train, y_train).mean(),
            "mean CV MAPE ": -cross_val_score(model, x_train, y_train, scoring=neg_mean_absolute_percentage_error).mean(),
            "mean CV MSE  ": -cross_val_score(model, x_train, y_train, scoring='neg_mean_squared_error').mean(),
            "mean CV RMSE ": -cross_val_score(model, x_train, y_train, scoring='neg_root_mean_squared_error').mean(),
            "mean CV MAE  ": -cross_val_score(model, x_train, y_train, scoring='neg_mean_absolute_error').mean(),
            "mean CV R2   ": cross_val_score(model, x_train, y_train, scoring='r2').mean(),
            "mean CV Explained Variance": cross_val_score(model, x_train, y_train, scoring='explained_variance').mean(),
            "mean CV MAX  ": -cross_val_score(model, x_train, y_train, scoring='max_error').mean()
            }
```

```python
def calculate_mape(y_true, y_pred):
    res = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return res
```

```python
def calculate_max_error(y_true, y_pred):
    res = max_error(y_true, y_pred)
    return res
```

```python
# Entfernen Sie die Zielvariable (Zz_Verkaufspreis)
data_without_target = df_modeling.drop('Zz_Verkaufspreis', axis=1)

# Erstellen Sie einen Imputer und ersetzen Sie fehlende Werte mit dem Median der jeweiligen Spalte
imputer = SimpleImputer(strategy='median')
imputed_data = imputer.fit_transform(data_without_target.select_dtypes(include=[np.number]))

# Skalierung mit allen Features
scaler_all_features = StandardScaler()
scaled_data_all_features = scaler_all_features.fit_transform(imputed_data)

# Ohne Bezirksfeatures (Lage)
scaler_without_district = StandardScaler()
data_without_district = data_without_target.drop('Lage', axis=1)
imputed_data_without_district = imputer.fit_transform(data_without_district.select_dtypes(include=[np.number]))
scaled_data_without_district = scaler_without_district.fit_transform(imputed_data_without_district)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia_values = []

# Testen Sie verschiedene Anzahlen von Clustern (z.B. von 1 bis 10)
num_clusters = range(1, 11)

for k in num_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_data_all_features)
    inertia_values.append(kmeans.inertia_)

# Plot der Inertia gegen die Anzahl der Cluster
plt.plot(num_clusters, inertia_values, 'bo-')
plt.xlabel('Anzahl der Cluster')
plt.ylabel('Inertia')
plt.title('Elbow-Methode zur Bestimmung der optimalen Anzahl von Clustern')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []

# Testen Sie verschiedene Anzahlen von Clustern (z.B. von 2 bis 10)
num_clusters = range(2, 11)

for k in num_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init= 10)
    cluster_labels = kmeans.fit_predict(scaled_data_all_features)
    silhouette_avg = silhouette_score(scaled_data_all_features, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot der Silhouetten-Koeffizienten gegen die Anzahl der Cluster
plt.plot(num_clusters, silhouette_scores, 'bo-')
plt.xlabel('Anzahl der Cluster')
plt.ylabel('Silhouetten-Koeffizient')
plt.title('Silhouetten-Koeffizient zur Bestimmung der optimalen Anzahl von Clustern')
plt.show()
```
